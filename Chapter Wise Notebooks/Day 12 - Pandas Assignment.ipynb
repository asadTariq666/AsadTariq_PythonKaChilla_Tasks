```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Importing Annual population of 2016-2018 from FAO.org

```python

pop = pd.read_csv('Annual population 2016-2018.csv')
print(pop.head())
```

## Change to Excel

```python
# install openpyxl
pop.to_excel('population.xlsx')
```

## Basic Statistics

```python
pop.describe()
```

## Dropping few Columns

```python
new_pop = pop.drop(['Domain','Flag','Item','Note'], axis=1) # axis1 = row number, axis2 =Column Number
new_pop
```

```python
new_pop.mean()
```

## groupBy (Pivot)

```python
# grouping by Area and gender

new_pop.groupby(['Area','Element']).mean()
```

```python
# grouping by Area, Year and gender

new_pop.groupby(['Area','Element','Year']).mean()
```

## Getting values of any column

```python
pop.value_counts('Year')
```

## Conditional retrieval

```python
# Check population greater than 200 million
pop[pop['Value']>200000].head()
```

```python
pop[pop['Value']>10000].groupby(['Area','Element','Year']).mean()
```
