## On Chilla Data

```python
chilla = pd.read_csv('ChillaData.csv')
chilla.head()
```

## Changing Binary optionset data to Numeric data

```python
field = chilla['field_of_study']
vaccinated = chilla['Vaccinated'].map({'Yes': 1, 'No': 0})
gender =chilla['Gender'].map({'Male': 1, 'Female': 0})
glasses = chilla['Glasses'].map({'Yes': 1, 'No': 0})
marital = chilla['Marital Status'].map({'Yes': 1, 'No': 0})
area = chilla['Area'].map({'Urbun': 1, 'Rural': 0})
```

## Scatterplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")


# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="First line of code", y="Height",
                hue=gender, size=vaccinated,
                hue_order=gender, color='red',
                sizes=(1, 8), linewidth=0,
                data=chilla, ax=ax)
```

## Lineplot

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(x="Age", y="Weight",
             hue=area, style=gender,
             data=chilla)
```

## Displot

```python
import seaborn as sns

sns.displot(
    chilla, x="Age", col=area, row=gender,
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
plt.figure(figsize=(18,6))
```

## Lmplot

```python
import seaborn as sns
sns.set_theme(style="ticks")



# Show the results of a linear regression within each dataset
sns.lmplot(x="Height", y="Weight", col="Age", hue="Age", data=chilla,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

## Violinplot

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=chilla, x="Age", y="Coding Hours", hue=gender,
               split=True, inner="quart", linewidth=1,
               )
sns.despine(left=True)
```

## Jointplot

```python
import numpy as np
import seaborn as sns
sns.set_theme(style="ticks")



sns.jointplot(data=chilla, x="Age", y="Height", color="#4CB391")
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    chilla,
    x="Height", hue="Network",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    log_scale=True,
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([500, 1000, 2000, 5000, 10000])
```

##  Boxplot and Stripplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("log")


# Plot the orbital period with horizontal boxes
sns.boxplot(x="Coding Hours", y="Location", data=chilla,
            whis=[0, 100], width=.6, palette="vlag")

# Add in points to show each observation
sns.stripplot(x="Coding Hours", y="Location", data=chilla,
              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

## Jointplot with kind="kde"

```python
import numpy as np
import seaborn as sns
sns.set_theme(style="ticks")



sns.jointplot(data=chilla, y="Weight", x="Height",kind="kde", color="#4CB391")
```
